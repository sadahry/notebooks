# Mel Scale

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker

def hz_to_mel(freqs):
    # original
    # ref. Stevens, S. S., John E. Volkmann and Edwin B. Newman. “A Scale for the Measurement of the Psychological Magnitude Pitch.” Journal of the Acoustical Society of America 8 (1937): 185-190.
    return 1127.0 * np.log(1.0 + freqs / 700.0)
    # librosa  
    # ref. Young, S., Evermann, G., Gales, M., Hain, T., Kershaw, D., Liu, X., Moore, G., Odell, J., Ollason, D., Povey, D., Valtchev, V., & Woodland, P. The HTK book, version 3.4. Cambridge University, March 2009.
    # return 2595.0 * np.log10(1.0 + freqs / 700.0)

freqs = np.linspace(1, 10000, num=1100)
mels = hz_to_mel(freqs)

plt.figure(figsize=(12, 4))
plt.plot(freqs, mels)
plt.xlabel('Frequency [Hz]')
plt.ylabel('Mel Scale [mel]')
plt.xticks(np.arange(0, 10001, 500.0))
plt.yticks(np.arange(0, 3001, 500.0))
plt.xlim(0, 10000)
plt.ylim(0, 3200)
plt.legend()
plt.grid()

plt.show()
```
