```python
import numpy as np
import matplotlib.pyplot as plt
fs = 44100
size = 4096
t = np.arange(0, size) / fs
frq = np.fft.fftfreq(size, 1/fs)

%matplotlib inline
```

```python
peaks = []
fset = np.linspace(0, 1000, 10001)
for f in fset:
	y = np.sin(2 * np.pi * f * t)
	y = y + np.sin(2 * 2 * np.pi * f * t)
	y = y + np.sin(3 * 2 * np.pi * f * t)
	y = y + np.sin(4 * 2 * np.pi * f * t)
	y = y + np.sin(5 * 2 * np.pi * f * t)
	Y = np.fft.fft(y)
	peaks.append(max(abs(Y)))
plt.plot(fset[1:3000], peaks[1:3000])
plt.show()
```
