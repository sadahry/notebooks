# GiNZA 係り受け
ref. https://www.nogawanogawa.com/entry/ginza-2

```python
import  ginza
import spacy
from spacy import displacy
```

```python
class DependencyAnalysis:
    def __init__(self):
        self.nlp = spacy.load('ja_ginza')
        ginza.set_split_mode(self.nlp, "A")
        
    def run(self, text):
        """係り受け解析"""
        doc = self.nlp(text)

        token_head_list = []
        
        for sent in doc.sents:
            for token in sent:
                token_head_list.append({"i":token.i, "orth":token.orth_, "base": token.lemma_, "head": token.head.i, "dep": token.dep_}) 

        displacy.render(doc, style='dep', jupyter=True, options={'distance': 90})

        return token_head_list
```

```python
dependency = DependencyAnalysis()
dependency.run(input("input text:")) 
```
